#### 1. Importing data

```python
#Importing orders_products_combined
df_ords_combined = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_combined.pkl'))
```

```python
#Importing products_checked
df_prods = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'products_checked.csv'))
```

#### 2. Checking output of data

For df_ords_combined:

```python
df_ords_combined.head()
```

```python
# Dropping unnecessary columns
df_ords_combined = df_ords_combined.drop(columns = ['Unnamed: 0.1', 'Unnamed: 0', '_merge'])
```

For df_prods:

```python
df_prods.head()
```

```python
# Dropping unnecessary column
df_prods = df_prods.drop(columns = ['Unnamed: 0'])
```

```python
df_prods.describe()
```

The best way to merge the data is by using the merge() function with the indicator parameter set to true.

```python
df_merged = df_ords_combined.merge(df_prods, on = 'product_id', indicator = True)
```

```python
#Confirming the results of the merge
df_merged.head()
```

```python
df_merged.shape
```

```python
#Confirming the results of the merge
df_merged['_merge'].value_counts()
```

```python
#Exporting ords_prods_merge
df_merged.to_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'ords_prods_merge.pkl'))
```
