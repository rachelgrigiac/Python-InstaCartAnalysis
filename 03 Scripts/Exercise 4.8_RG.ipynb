2. Finding the aggregated mean of the 'user_order_number' column grouped by 'department_id':

```python
ords_prods_merge.head()
```

```python
ords_prods_merge.groupby('department_id').agg({'user_order_number': ['mean']})
```

3. The results now include all the departments, providing more accurate information about which departments have more order numbers per user. Department #21 has the highest number, which was not included in the subset.

4. Creating a loyalty flag for existing customers:

```python
# Create new column 'max_order'
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['user_order_number'].transform(np.max)
```

```python
# Check the results
ords_prods_merge.head(15)
```

```python
# Deriving colum with loc()
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Check the results
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Check the right flags were assigned
ords_prods_merge[['user_id','loyalty_flag','user_order_number']].head(30)
```

5. Do the prices of products purchased by loyal customers differ from those purchased by regular or new customers?

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices':['mean','min','max']})
```

6. Different types of spenders:

```python
# Create new column 'average_price'
ords_prods_merge['average_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Deriving colum with loc()
ords_prods_merge.loc[ords_prods_merge['average_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Check the results
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# Check the right flags were assigned
ords_prods_merge[['user_id','spending_flag','average_price']].head(30)
```

```python
ords_prods_merge[['user_id','spending_flag','average_price']].tail(60)
```

7. Determine frequent versus non-frequent customers:

```python
# Create new column 'average_price'
ords_prods_merge['order_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Deriving new column
ords_prods_merge.loc[ords_prods_merge['order_frequency'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['order_frequency'] > 10) & (ords_prods_merge['order_frequency'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['order_frequency'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Check the results
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```
