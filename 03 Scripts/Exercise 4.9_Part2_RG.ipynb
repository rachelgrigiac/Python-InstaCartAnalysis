```python
ords_prods_cust = pd.read_pickle(os.path.join( path, '02 Data', 'Prepared Data', 'ords_prods_cust.pkl'))
```

## 3.

```python
hist = ords_prods_cust['order_hour_of_day'].plot.hist(bins = 24)
```

Customers usually prefer to order in the late morning or afternoon. Orders start increasing from 5am, reaching a peak between 9am to 4pm. Most orders are placed between 9am to 10am. After 4pm, orders decrease. 
It is also less likely for people to order between midnight and 7am.

## 4.

```python
bar = ords_prods_cust['loyalty_flag'].value_counts().plot.bar(color = ['lightblue', 'gold', 'orange'])
```

## 5.

```python
# Randomize data before sampling
np.random.seed(4)
```

```python
# Create a dev list (70/30)
dev = np.random.rand(len(ords_prods_cust)) <= 0.7
```

```python
# Split data in two samples
big = ords_prods_cust[dev]
small = ords_prods_cust [~dev]
```

```python
# Check the results
len(ords_prods_cust)
```

```python
len(big) + len(small)
```

```python
# Reduce to only necessary columns
df2 = small[['prices','order_hour_of_day']]
```

```python
line = sns.lineplot(data = df2, x='order_hour_of_day', y= 'prices')
```

## 6.

```python
# Reduce to only necessary columns
df3 = small[['age','n_dependants']]
```

```python
line2 = sns.lineplot(data = df3, x ='age', y ='n_dependants')
```

There seems to be no pattern between age and the number of dependents.

## 7.

```python
scat = sns.scatterplot(x = 'age', y = 'income', data = small)
```

Before age 40, most points cluster below 200000 in income, with fewer points between 200000 and 400000. After age 40, while the majority still earn below 200000, there is a notable increase in points between 200000 and 300000, and the income range extends up to 600000.
